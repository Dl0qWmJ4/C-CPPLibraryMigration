```python
reason_type = {
    '0': 'none',
    '1': 'simplification', 
    '2': 'integration', 
    '3': 'feature', 
    '4': 'usability', 
    '5': 'performance',
    '6': 'deprecation',  
    '7': 'bug or issue',
    '8': 'organization',  
    '9': 'license',
    '10': 'security',
    '11': 'popularity',
    '13': 'new',  
    '14': 'distribution or install error',
    '15': 'abandon support',
    '16': 'size',
    '17': 'upgrade',
    '18': 'activity',
}
```

```python
import pandas as pd
reason = pd.read_csv('dataset/c_reason.csv')

def get_reason_type(value):
    return reason_type[str(value)]

reason['reason_type'] = reason['reason'].apply(get_reason_type)
reason_dist = reason['reason_type'].value_counts().reset_index()
reason_dist.columns = ['reason', 'num']
reason_dist['proportion'] = (reason_dist['num'] * 10000.0 / len(reason)).astype('int') / 100
reason_dist
```
